# Análisis Exploratorio de Datos - Vehículos en Venta

Este notebook contiene visualizaciones interactivas utilizando Plotly para explorar el conjunto de datos de anuncios de venta de vehículos.

```python
# Importar librerías necesarias
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
# Leer los datos del archivo CSV
car_data = pd.read_csv('../upload/vehicles_us.csv')

# Mostrar las primeras filas del dataset
car_data.head()
```

```python
# Información general del dataset
car_data.info()
```

```python
# Estadísticas descriptivas
car_data.describe()
```

## Histograma: Distribución del Odómetro

El histograma muestra la distribución del kilometraje (odómetro) de los vehículos en el dataset.

```python
# Crear un histograma utilizando plotly.graph_objects
fig = go.Figure(data=[go.Histogram(x=car_data['odometer'], nbinsx=50)])

# Actualizar el diseño del gráfico
fig.update_layout(
    title_text='Distribución del Odómetro',
    xaxis_title='Odómetro (millas)',
    yaxis_title='Frecuencia',
    bargap=0.1
)

fig.show()
```

## Gráfico de Dispersión: Relación entre Odómetro y Precio

Este gráfico muestra la relación entre el kilometraje del vehículo y su precio.

```python
# Crear un scatter plot utilizando plotly.graph_objects
fig = go.Figure(data=[go.Scatter(
    x=car_data['odometer'],
    y=car_data['price'],
    mode='markers',
    marker=dict(
        size=5,
        color=car_data['price'],
        colorscale='Viridis',
        showscale=True
    ),
    opacity=0.6
)])

# Actualizar el diseño del gráfico
fig.update_layout(
    title_text='Relación entre Odómetro y Precio',
    xaxis_title='Odómetro (millas)',
    yaxis_title='Precio (USD)'
)

fig.show()
```

## Histograma: Distribución de Precios

Análisis de la distribución de precios de los vehículos.

```python
# Filtrar precios razonables (eliminar outliers extremos)
filtered_data = car_data[(car_data['price'] > 0) & (car_data['price'] < 100000)]

# Crear histograma de precios
fig = go.Figure(data=[go.Histogram(x=filtered_data['price'], nbinsx=50)])

fig.update_layout(
    title_text='Distribución de Precios de Vehículos',
    xaxis_title='Precio (USD)',
    yaxis_title='Frecuencia',
    bargap=0.1
)

fig.show()
```

## Gráfico de Barras: Distribución por Condición del Vehículo

```python
# Contar vehículos por condición
condition_counts = car_data['condition'].value_counts().reset_index()
condition_counts.columns = ['condition', 'count']

# Crear gráfico de barras
fig = go.Figure(data=[go.Bar(
    x=condition_counts['condition'],
    y=condition_counts['count'],
    marker_color='steelblue'
)])

fig.update_layout(
    title_text='Distribución de Vehículos por Condición',
    xaxis_title='Condición',
    yaxis_title='Cantidad de Vehículos'
)

fig.show()
```

## Gráfico de Barras: Distribución por Tipo de Vehículo

```python
# Contar vehículos por tipo
type_counts = car_data['type'].value_counts().reset_index()
type_counts.columns = ['type', 'count']

# Crear gráfico de barras
fig = go.Figure(data=[go.Bar(
    x=type_counts['type'],
    y=type_counts['count'],
    marker_color='coral'
)])

fig.update_layout(
    title_text='Distribución de Vehículos por Tipo',
    xaxis_title='Tipo de Vehículo',
    yaxis_title='Cantidad'
)

fig.show()
```

## Histograma: Distribución del Año del Modelo

```python
# Filtrar años válidos
valid_years = car_data[car_data['model_year'].notna()]

# Crear histograma de años
fig = go.Figure(data=[go.Histogram(x=valid_years['model_year'], nbinsx=30)])

fig.update_layout(
    title_text='Distribución del Año del Modelo',
    xaxis_title='Año del Modelo',
    yaxis_title='Frecuencia',
    bargap=0.1
)

fig.show()
```

## Gráfico de Dispersión: Precio vs Año del Modelo

```python
# Filtrar datos con año y precio válidos
filtered_data = car_data[
    (car_data['model_year'].notna()) & 
    (car_data['price'] > 0) & 
    (car_data['price'] < 100000)
]

# Crear scatter plot
fig = go.Figure(data=[go.Scatter(
    x=filtered_data['model_year'],
    y=filtered_data['price'],
    mode='markers',
    marker=dict(
        size=5,
        color=filtered_data['model_year'],
        colorscale='RdYlGn',
        showscale=True
    ),
    opacity=0.6
)])

fig.update_layout(
    title_text='Relación entre Año del Modelo y Precio',
    xaxis_title='Año del Modelo',
    yaxis_title='Precio (USD)'
)

fig.show()
```

## Box Plot: Precio por Tipo de Vehículo

```python
# Usar plotly express para box plot
fig = px.box(
    car_data[car_data['price'] < 100000],
    x='type',
    y='price',
    title='Distribución de Precios por Tipo de Vehículo'
)

fig.update_layout(
    xaxis_title='Tipo de Vehículo',
    yaxis_title='Precio (USD)'
)

fig.show()
```
